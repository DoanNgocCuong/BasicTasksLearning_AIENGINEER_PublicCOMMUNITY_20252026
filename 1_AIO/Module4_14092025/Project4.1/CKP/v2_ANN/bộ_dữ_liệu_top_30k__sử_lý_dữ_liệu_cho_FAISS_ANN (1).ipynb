```python
import ast
import pandas as pd # Giả sử bạn đang dùng pandas

# Hàm để chuyển đổi an toàn, chỉ áp dụng literal_eval nếu là chuỗi
def safe_literal_eval(data):
    if isinstance(data, str):
        try:
            return ast.literal_eval(data)
        except (ValueError, SyntaxError):
            # Trả về một danh sách rỗng hoặc xử lý lỗi nếu chuỗi không hợp lệ
            return []
    # Nếu đã là list hoặc kiểu dữ liệu khác, trả về nguyên bản
    return data

# BƯỚC MỚI VÀ QUAN TRỌNG: Áp dụng hàm chuyển đổi an toàn
df['parent_labels'] = df['parent_labels'].apply(safe_literal_eval)

# 3. San phẳng (flatten) dữ liệu và lấy các nhãn duy nhất
all_parent_labels_flat = [label for sublist in df['parent_labels'] for label in sublist if label and label.isalnum()]
unique_parent_labels = sorted(list(set(all_parent_labels_flat)))

print("\nĐã lấy xong các nhãn 'parent_labels' duy nhất.")
print(f"Danh sách các nhãn duy nhất: {unique_parent_labels}")

# 4. Tạo embeddings cho các nhãn và chuẩn hóa
print("\nĐang tạo embeddings cho các 'parent_labels' và chuẩn hóa...")
parent_label_embeddings_list = model.encode(
    unique_parent_labels,
    show_progress_bar=True,
    normalize_embeddings=True # Thêm tham số này để chuẩn hóa vector
)
print("Đã tạo xong và chuẩn hóa embeddings.")

# 5. Tạo dictionary ánh xạ nhãn với embeddings đã chuẩn hóa
parent_label_dict = {name: embedding for name, embedding in zip(unique_parent_labels, parent_label_embeddings_list)}
print("\nĐã tạo xong dictionary cho parent_labels.")
print(f"Số lượng entries trong dictionary: {len(parent_label_dict)}")
```

```python
import numpy as np
import faiss

d = X_train.shape[1]
# Dùng IndexFlatL2 cho độ chính xác cao nhất, hoặc IndexIVFPQ cho tốc độ
index = faiss.IndexFlatL2(d)
index.add(X_train.astype('float32'))

print(f"Đã xây dựng index với {index.ntotal} vector từ tập training.")

#### Giai đoạn 3: Logic Dự đoán (Prediction)
from collections import Counter
import itertools

def predict_labels(query_vector, k=10):
    # Bước 1: Tìm k hàng xóm
    distances, indices = index.search(np.array([query_vector]).astype('float32'), k)
    neighbor_indices = indices[0]

    # Bước 2: Lấy nhãn của hàng xóm
    neighbor_parent_labels = [y_train_parent[i] for i in neighbor_indices]
    neighbor_child_labels = [y_train_child[i] for i in neighbor_indices]

    # Bước 3: Voting cho nhãn cha
    # Flatten list of lists và đếm
    parent_counts = Counter(itertools.chain.from_iterable(neighbor_parent_labels))
    # Dự đoán nhãn cha nào xuất hiện ít nhất 2 lần (có thể điều chỉnh ngưỡng này)
    predicted_parents = [label for label, count in parent_counts.items() if count >= 2]

    # Voting cho nhãn con
    child_counts = Counter(itertools.chain.from_iterable(neighbor_child_labels))
    predicted_children = [label for label, count in child_counts.items() if count >= 2]

    return predicted_parents, predicted_children

# Ví dụ: dự đoán cho abstract đầu tiên trong tập test
predicted_p, predicted_c = predict_labels(X_test[0])
print(f"Nhãn cha thực tế: {y_test_parent[0]}")
print(f"Nhãn cha dự đoán: {predicted_p}")
```

```python
from collections import Counter
import itertools
import ast
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score, hamming_loss
from tqdm import tqdm # Thư viện để hiển thị thanh tiến trình đẹp mắt
import faiss

# Hàm an toàn để chuyển đổi chuỗi thành list
def safe_literal_eval(data):
    if isinstance(data, str):
        try:
            return ast.literal_eval(data)
        except (ValueError, SyntaxError):
            # Điểm sửa chữa: Tách chuỗi nhãn con bằng dấu cách
            return data.split()
    elif isinstance(data, list):
        return data
    else:
        return []

# Hàm dự đoán đã được sửa lỗi
def predict_labels_hierarchical_v2(query_vector, index, y_train_parent, y_train_child, k=15, max_parents=2, max_children_per_parent=1):
    distances, indices = index.search(np.array([query_vector]).astype('float32'), k)
    neighbor_indices = indices[0]

    neighbor_parent_labels = [safe_literal_eval(y_train_parent[i]) for i in neighbor_indices]
    parent_counts = Counter(itertools.chain.from_iterable(neighbor_parent_labels))
    top_parents = [label for label, count in parent_counts.most_common(max_parents)]

    final_predictions = {}

    for parent in top_parents:
        valid_child_labels_for_voting = []
        for i in neighbor_indices:
            neighbor_p = safe_literal_eval(y_train_parent[i])
            # Sử dụng hàm đã sửa để tách chuỗi nhãn con
            neighbor_c = safe_literal_eval(y_train_child[i])

            if parent in neighbor_p:
                valid_child_labels_for_voting.extend(neighbor_c)

        if not valid_child_labels_for_voting:
            final_predictions[parent] = None
            continue

        child_counts = Counter(valid_child_labels_for_voting)
        if child_counts:
            top_child = child_counts.most_common(max_children_per_parent)[0][0]
            final_predictions[parent] = top_child
        else:
            final_predictions[parent] = None

    predicted_parents = list(final_predictions.keys())
    predicted_children = [child for child in final_predictions.values() if child is not None]
    return predicted_parents, predicted_children


# --- BƯỚC 3: TÍNH TOÁN VÀ IN KẾT QUẢ ĐÁNH GIÁ ---

def display_evaluation_report(y_true_bin, y_pred_bin, label_type):
    """Hàm để in báo cáo đánh giá chi tiết."""
    print(f"\n--- BÁO CÁNH GIÁ CHO: {label_type.upper()} ---")

    # Accuracy (Subset Accuracy): Tỷ lệ dự đoán đúng *chính xác* toàn bộ tập nhãn.
    # Đây là chỉ số rất nghiêm ngặt.
    accuracy = accuracy_score(y_true_bin, y_pred_bin)
    print(f"Accuracy (Subset Accuracy): {accuracy:.4f}")

    # Hamming Loss: Tỷ lệ phần trăm nhãn bị dự đoán sai trên tổng số nhãn.
    # Càng thấp càng tốt.
    h_loss = hamming_loss(y_true_bin, y_pred_bin)
    print(f"Hamming Loss (lower is better): {h_loss:.4f}")


    print("\n--- Các chỉ số trung bình ---")
    # 'samples': Tính chỉ số cho từng mẫu rồi lấy trung bình. Rất hữu ích.
    precision_samples = precision_score(y_true_bin, y_pred_bin, average='samples', zero_division=0)
    recall_samples = recall_score(y_true_bin, y_pred_bin, average='samples', zero_division=0)
    f1_samples = f1_score(y_true_bin, y_pred_bin, average='samples', zero_division=0)


    print(f"Precision (per-sample average): {precision_samples:.4f}")
    print(f"Recall (per-sample average):    {recall_samples:.4f}")
    print(f"F1-Score (per-sample average):  {f1_samples:.4f}")

    # 'micro': Tính toán trên toàn cục bằng cách đếm tổng số true positives, false negatives và false positives.
    f1_micro = f1_score(y_true_bin, y_pred_bin, average='micro', zero_division=0)
    print(f"\nF1-Score (micro average): {f1_micro:.4f}")

    # 'macro': Tính chỉ số cho từng nhãn rồi lấy trung bình (không quan tâm đến sự mất cân bằng).
    f1_macro = f1_score(y_true_bin, y_pred_bin, average='macro', zero_division=0)
    print(f"F1-Score (macro average): {f1_macro:.4f}")


# ----------------- Phần chạy đánh giá -----------------

# Đảm bảo các biến abstract_embeddings, original_parent_labels, original_children_labels đã được load
# Chia dữ liệu thành tập train và test
X_train, X_test, y_train_parent, y_test_parent, y_train_child, y_test_child = train_test_split(
    abstract_embeddings,
    original_parent_labels,
    original_children_labels,
    test_size=0.2,
    random_state=42
)

# Xây dựng Faiss index
d = X_train.shape[1]
index = faiss.IndexFlatL2(d)
index.add(X_train.astype('float32'))
print(f"Đã xây dựng index với {index.ntotal} vector từ tập training.")


# Chuẩn bị binarizer
all_parent_labels = y_train_parent + y_test_parent
all_child_labels = y_train_child + y_test_child

parent_binarizer = MultiLabelBinarizer()
parent_binarizer.fit(all_parent_labels)

child_binarizer = MultiLabelBinarizer()
child_binarizer.fit(all_child_labels)

y_test_parent_bin = parent_binarizer.transform(y_test_parent)
y_test_child_bin = child_binarizer.transform(y_test_child)


all_predicted_parents_fixed = []
all_predicted_children_fixed = []

print(f"Bắt đầu chạy dự đoán trên {len(X_test)} mẫu của tập test với hàm đã sửa lỗi...")
# Sử dụng tqdm để hiển thị thanh tiến trình
for i in tqdm(range(len(X_test))):
    # Pass the required arguments to the function
    pred_p, pred_c = predict_labels_hierarchical_v2(X_test[i], index, y_train_parent, y_train_child)
    all_predicted_parents_fixed.append(pred_p)
    all_predicted_children_fixed.append(pred_c) # Lưu cả nhãn con dự đoán nếu cần sau này


print("Đã hoàn tất dự đoán với hàm đã sửa lỗi.")

y_pred_parent_bin_fixed = parent_binarizer.transform(all_predicted_parents_fixed)
y_pred_child_bin_fixed = child_binarizer.transform(all_predicted_children_fixed)


# Chạy báo cáo cho nhãn cha
display_evaluation_report(y_test_parent_bin, y_pred_parent_bin_fixed, "Nhãn Cha")

# Chạy báo cáo cho nhãn con
display_evaluation_report(y_test_child_bin, y_pred_child_bin_fixed, "Nhãn Con")
```
