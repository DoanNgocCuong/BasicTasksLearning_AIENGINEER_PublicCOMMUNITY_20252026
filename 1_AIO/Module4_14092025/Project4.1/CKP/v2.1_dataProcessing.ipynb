```python
from datasets import load_dataset
import glob
import os

# Cách 1: Sử dụng đường dẫn đầy đủ đến thư mục chứa file .arrow
# Định nghĩa đường dẫn đến thư mục chứa các file .arrow
arrow_dir = "dataset/UniverseTBD___arxiv-abstracts-large/default/0.0.0/6020a62078a73d7ca02b86a4a775af7caba42d5e"

# Tìm tất cả file .arrow trong thư mục đó
data_files = glob.glob(os.path.join(arrow_dir, "*.arrow"))

# In ra để kiểm tra
print(f"Số lượng file .arrow tìm thấy: {len(data_files)}")
print("Danh sách file:")
for file in data_files:
    print(f"  - {file}")

# Load dataset từ các file .arrow
if data_files:  # Kiểm tra xem có file nào không
    full_dataset = load_dataset("arrow", data_files=data_files)
    print(f"\nDataset đã được load thành công!")
    print(f"Số lượng samples: {len(full_dataset['train'])}")
else:
    print("Không tìm thấy file .arrow nào!")
```

## categories statistic :


```python

# Code tối ưu nhất - In 100 categories phổ biến nhất
print("=== THỐNG KÊ 100 CATEGORIES PHỔ BIẾN NHẤT ===")

import pandas as pd
import time

# Bắt đầu đo thời gian
start_time = time.time()

print("🔄 Đang chuyển dataset sang pandas...")

# Chuyển dataset sang pandas (tối ưu cho 32GB RAM)
df = full_dataset['train'].to_pandas()

print(f"✅ Hoàn thành trong {time.time() - start_time:.1f} giây!")

# Thống kê categories (chỉ lấy top 100)
category_stats = df['categories'].value_counts().head(100)

# Hiển thị kết quả
print(f"\n�� TOP 100 CATEGORIES PHỔ BIẾN NHẤT:")
print(f"Tổng số bài báo: {len(df):,}")
print(f"Tổng thời gian: {time.time() - start_time:.1f} giây")
print("-" * 70)

# In ra 100 categories
for i, (category, count) in enumerate(category_stats.items(), 1):
    percentage = (count / len(df)) * 100
    print(f"{i:3d}. {category:<30} | {count:>8,} bài ({percentage:5.2f}%)")
```

```python
# ===== LÀM SẠCH CATEGORIES - LOẠI BỎ PHẦN SAU DẤU "." =====
print("=== LÀM SẠCH CATEGORIES ===")

import pandas as pd
import time

# Bắt đầu đo thời gian
start_time = time.time()

print("🔄 Đang làm sạch categories...")

# Tạo bản sao của dataframe để không ảnh hưởng đến dữ liệu gốc
df_cleaned = df.copy()

# Hiển thị một vài ví dụ trước khi làm sạch
print("\n📋 VÍ DỤ TRƯỚC KHI LÀM SẠCH:")
sample_categories = df_cleaned['categories'].head(10).tolist()
for i, cat in enumerate(sample_categories, 1):
    print(f"  {i:2d}. {cat}")

# Loại bỏ phần sau dấu "." trong cột categories
# Sử dụng str.split('.') để tách chuỗi tại dấu "."
# Sau đó lấy phần đầu tiên [0] (trước dấu ".")
df_cleaned['categories_cleaned'] = df_cleaned['categories'].str.split('.').str[0]

# Hiển thị một vài ví dụ sau khi làm sạch
print("\n✨ VÍ DỤ SAU KHI LÀM SẠCH:")
sample_cleaned = df_cleaned['categories_cleaned'].head(10).tolist()
for i, cat in enumerate(sample_cleaned, 1):
    print(f"  {i:2d}. {cat}")

print(f"\n✅ Hoàn thành làm sạch trong {time.time() - start_time:.1f} giây!")
```

```python
# ===== THỐNG KÊ CATEGORIES SAU KHI LÀM SẠCH =====
print("=== THỐNG KÊ CATEGORIES SAU KHI LÀM SẠCH ===")

# Thống kê categories đã làm sạch (lấy top 50)
cleaned_category_stats = df_cleaned['categories_cleaned'].value_counts().head(1000)

# Hiển thị kết quả
print(f"\n📊 TOP 50 CATEGORIES SAU KHI LÀM SẠCH:")
print(f"Tổng số bài báo: {len(df_cleaned):,}")
print(f"Tổng thời gian: {time.time() - start_time:.1f} giây")
print("-" * 70)

# In ra 50 categories đã làm sạch
for i, (category, count) in enumerate(cleaned_category_stats.items(), 1):
    percentage = (count / len(df_cleaned)) * 100
    print(f"{i:2d}. {category:<20} | {count:>8,} bài ({percentage:5.2f}%)")

# So sánh số lượng categories trước và sau khi làm sạch
original_unique = df['categories'].nunique()
cleaned_unique = df_cleaned['categories_cleaned'].nunique()

print(f"\n📈 SO SÁNH:")
print(f"  - Số categories gốc: {original_unique:,}")
print(f"  - Số categories sau làm sạch: {cleaned_unique:,}")
print(f"  - Giảm: {original_unique - cleaned_unique:,} categories ({((original_unique - cleaned_unique) / original_unique * 100):.1f}%)")
```

```python
# ===== LỌC DỮ LIỆU THEO TOP 40 CATEGORIES =====
print("=== LỌC DỮ LIỆU THEO TOP 40 CATEGORIES ===")

import pandas as pd
import time

# Bắt đầu đo thời gian
start_time = time.time()

print("🔄 Đang lọc dữ liệu theo top 40 categories...")

# 1. Lấy top 40 categories phổ biến nhất (sau khi làm sạch)
top_40_categories = df_cleaned['categories_cleaned'].value_counts().head(40)

print(f"\n📊 TOP 40 CATEGORIES ĐƯỢC CHỌN:")
print("Hạng | Category           | Số bài báo | Tỷ lệ")
print("-" * 55)

for i, (category, count) in enumerate(top_40_categories.items(), 1):
    percentage = (count / len(df_cleaned)) * 100
    print(f"{i:3d}  | {category:<18} | {count:>8,} | {percentage:5.2f}%")

# 2. Lọc dataset chỉ giữ lại các bài báo thuộc top 40 categories
print(f"\n🔍 Đang lọc dataset...")

# Tạo mask để lọc dữ liệu
mask = df_cleaned['categories_cleaned'].isin(top_40_categories.index)

# Tạo dataset mới
df_top40 = df_cleaned[mask].copy()

print(f"✅ Hoàn thành lọc dữ liệu trong {time.time() - start_time:.1f} giây!")

# 3. Thống kê dataset mới
print(f"\n📈 THỐNG KÊ DATASET MỚI:")
print(f"  - Số bài báo gốc: {len(df_cleaned):,}")
print(f"  - Số bài báo sau lọc: {len(df_top40):,}")
print(f"  - Tỷ lệ giữ lại: {(len(df_top40)/len(df_cleaned)*100):.1f}%")
print(f"  - Số categories: {df_top40['categories_cleaned'].nunique()}")

# 4. Kiểm tra phân bố dữ liệu
print(f"\n📊 PHÂN BỐ DỮ LIỆU TRONG DATASET MỚI:")
category_distribution = df_top40['categories_cleaned'].value_counts()

print("Hạng | Category           | Số bài báo | Tỷ lệ trong dataset mới")
print("-" * 70)

for i, (category, count) in enumerate(category_distribution.items(), 1):
    percentage = (count / len(df_top40)) * 100
    print(f"{i:3d}  | {category:<18} | {count:>8,} | {percentage:5.2f}%")
```

```python
# ===== KIỂM TRA CHẤT LƯỢNG DATASET MỚI =====
print("=== KIỂM TRA CHẤT LƯỢNG DATASET MỚI ===")

# 1. Kiểm tra cấu trúc dataset mới
print("🔍 1. CẤU TRÚC DATASET MỚI:")
print(f"  - Kích thước: {df_top40.shape}")
print(f"  - Các cột: {list(df_top40.columns)}")

# 2. Kiểm tra dữ liệu thiếu
print(f"\n🔍 2. KIỂM TRA DỮ LIỆU THIẾU:")
missing_data = df_top40.isnull().sum()
for col, missing_count in missing_data.items():
    if missing_count > 0:
        percentage = (missing_count / len(df_top40)) * 100
        print(f"  - {col}: {missing_count:,} ({percentage:.2f}%)")

# 3. Hiển thị một vài ví dụ dữ liệu
print(f"\n🔍 3. VÍ DỤ DỮ LIỆU TRONG DATASET MỚI:")
print("Index | Categories (gốc)     | Categories (cleaned) | Title (50 ký tự đầu)")
print("-" * 90)

# Lấy 10 ví dụ đầu tiên
sample_data = df_top40[['categories', 'categories_cleaned', 'title']].head(10)
for idx, row in sample_data.iterrows():
    title_short = row['title'][:50] + "..." if len(row['title']) > 50 else row['title']
    print(f"{idx:5d} | {row['categories']:<20} | {row['categories_cleaned']:<18} | {title_short}")

# 4. Thống kê độ dài abstract
print(f"\n🔍 4. THỐNG KÊ ĐỘ DÀI ABSTRACT:")
if 'abstract' in df_top40.columns:
    abstract_lengths = df_top40['abstract'].str.len()
    print(f"  - Độ dài trung bình: {abstract_lengths.mean():.0f} ký tự")
    print(f"  - Độ dài ngắn nhất: {abstract_lengths.min():,} ký tự")
    print(f"  - Độ dài dài nhất: {abstract_lengths.max():,} ký tự")
    print(f"  - Độ lệch chuẩn: {abstract_lengths.std():.0f} ký tự")

# 5. Lưu dataset mới
print(f"\n💾 5. LƯU DATASET MỚI:")
print("  - Dataset mới được lưu trong biến 'df_top40'")
print("  - Bạn có thể sử dụng df_top40 cho các bước tiếp theo")
print("  - Kích thước nhỏ hơn, dễ xử lý hơn cho machine learning")
```

```python
# ===== LƯU DATASET MỚI VÀO FILE =====
print("=== LƯU DATASET MỚI VÀO FILE ===")

import os

# 1. Tạo thư mục để lưu dataset mới
output_dir = "filtered_datasets"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
    print(f"✅ Đã tạo thư mục: {output_dir}")

# 2. Lưu dataset mới dưới dạng CSV
output_file = os.path.join(output_dir, "arxiv_top40_categories.csv")
print(f"🔄 Đang lưu dataset vào: {output_file}")

# Lưu file CSV
df_top40.to_csv(output_file, index=False, encoding='utf-8')

print(f"✅ Đã lưu dataset thành công!")
print(f"  - File: {output_file}")
print(f"  - Kích thước: {df_top40.shape}")
print(f"  - Dung lượng file: {os.path.getsize(output_file) / (1024*1024):.1f} MB")

# 3. Tạo file thống kê
stats_file = os.path.join(output_dir, "dataset_statistics.txt")
print(f"\n📊 Đang tạo file thống kê: {stats_file}")

with open(stats_file, 'w', encoding='utf-8') as f:
    f.write("=== THỐNG KÊ DATASET TOP 40 CATEGORIES ===\n\n")
    f.write(f"Tổng số bài báo: {len(df_top40):,}\n")
    f.write(f"Số categories: {df_top40['categories_cleaned'].nunique()}\n")
    f.write(f"Tỷ lệ giữ lại từ dataset gốc: {(len(df_top40)/len(df_cleaned)*100):.1f}%\n\n")
    
    f.write("TOP 40 CATEGORIES:\n")
    f.write("-" * 50 + "\n")
    for i, (category, count) in enumerate(category_distribution.items(), 1):
        percentage = (count / len(df_top40)) * 100
        f.write(f"{i:2d}. {category:<18} | {count:>8,} bài ({percentage:5.2f}%)\n")

print(f"✅ Đã tạo file thống kê thành công!")

# 4. Hiển thị thông tin cuối cùng
print(f"\n🎉 HOÀN THÀNH LỌC DỮ LIỆU!")
print("=" * 50)
print(f"📁 Thư mục output: {output_dir}")
print(f"📄 File dữ liệu: arxiv_top40_categories.csv")
print(f"📊 File thống kê: dataset_statistics.txt")
print(f"🔢 Số bài báo: {len(df_top40):,}")
print(f"🏷️  Số categories: {df_top40['categories_cleaned'].nunique()}")
print(f"💾 Biến trong notebook: df_top40")
```
