````python
import ast
import pandas as pd # Giả sử bạn đang dùng pandas

# Hàm để chuyển đổi an toàn, chỉ áp dụng literal_eval nếu là chuỗi
def safe_literal_eval(data):
    if isinstance(data, str):
        try:
            return ast.literal_eval(data)
        except (ValueError, SyntaxError):
            # Trả về một danh sách rỗng hoặc xử lý lỗi nếu chuỗi không hợp lệ
            return []
    # Nếu đã là list hoặc kiểu dữ liệu khác, trả về nguyên bản
    return data

# BƯỚC MỚI VÀ QUAN TRỌNG: Áp dụng hàm chuyển đổi an toàn
df['parent_labels'] = df['parent_labels'].apply(safe_literal_eval)

# 3. San phẳng (flatten) dữ liệu và lấy các nhãn duy nhất
all_parent_labels_flat = [label for sublist in df['parent_labels'] for label in sublist if label and label.isalnum()]
unique_parent_labels = sorted(list(set(all_parent_labels_flat)))

print("\nĐã lấy xong các nhãn 'parent_labels' duy nhất.")
print(f"Danh sách các nhãn duy nhất: {unique_parent_labels}")

# 4. Tạo embeddings cho các nhãn và chuẩn hóa
print("\nĐang tạo embeddings cho các 'parent_labels' và chuẩn hóa...")
parent_label_embeddings_list = model.encode(
    unique_parent_labels,
    show_progress_bar=True,
    normalize_embeddings=True # Thêm tham số này để chuẩn hóa vector
)
print("Đã tạo xong và chuẩn hóa embeddings.")

# 5. Tạo dictionary ánh xạ nhãn với embeddings đã chuẩn hóa
parent_label_dict = {name: embedding for name, embedding in zip(unique_parent_labels, parent_label_embeddings_list)}
print("\nĐã tạo xong dictionary cho parent_labels.")
print(f"Số lượng entries trong dictionary: {len(parent_label_dict)}")
````

````python
import pandas as pd
import ast
import json

# ===== Cấu hình =====
data_path = "/content/drive/MyDrive/AIO/top 30 main categories.csv"  # Đường dẫn file trên Google Drive
out_path  = "processed_arxiv_data.csv" # Tên file output (sẽ lưu trong thư mục hiện tại của Colab)
# Điều chỉnh cột theo yêu cầu: categories là con, main_categories là cha
cols = ["authors", "title", "abstract", "categories", "main_categories"]

# ===== Đọc file chỉ lấy cột cần =====
df = pd.read_csv(data_path, usecols=cols)

# # Đổi tên cột theo yêu cầu - Bỏ qua bước đổi tên cột 'categories'
df = df.rename(columns={
    'main_categories': 'parent_labels'
})

# ===== Chuẩn hoá 2 cột list (nếu đang lưu dạng chuỗi list trong CSV) =====
# Cần kiểm tra kiểu dữ liệu trước khi áp dụng ast.literal_eval
def safe_literal_eval(data, col_name):
    if isinstance(data, str):
        try:
            return ast.literal_eval(data)
        except (ValueError, SyntaxError):
            # Nếu không phải chuỗi list hợp lệ, thử tách bằng dấu cách cho nhãn con ('categories')
            # Chú ý: Đây có thể cần điều chỉnh tùy vào format dữ liệu gốc không phải list string
            return data.split() if col_name == 'categories' and isinstance(data, str) else []
    elif isinstance(data, list):
        return data # Nếu đã là list, giữ nguyên
    else:
        return [] # Xử lý các kiểu dữ liệu khác (ví dụ: NaN)

# Áp dụng hàm chuẩn hóa cho cả hai cột, truyền tên cột để xử lý 'categories' khác
df['parent_labels'] = df['parent_labels'].apply(lambda x: safe_literal_eval(x, 'parent_labels'))
df['categories'] = df['categories'].apply(lambda x: safe_literal_eval(x, 'categories'))


# ===== Tạo cột json_categories: với mỗi cha, quét toàn bộ con để lấy con có tiền tố đúng =====
# Sử dụng tên cột 'categories' (nhãn con)
df["json_categories"] = df.apply(
    lambda r: {
        p: [c for c in (r["categories"] or [])
             if isinstance(c, str) and (c == p or c.startswith(p + ".") or c.startswith(p + "-"))]
        for p in (r["parent_labels"] or [])
    },
    axis=1
)

# Tuỳ chọn: nếu muốn lưu ra CSV ở dạng chuỗi JSON cho cột json_categories
df["json_categories"] = df["json_categories"].apply(lambda d: json.dumps(d, ensure_ascii=False))

# ===== Lưu kết quả =====
# Giữ lại các cột cần thiết sau khi xử lý, sử dụng 'categories' cho nhãn con
df_processed = df[['authors', 'title', 'abstract', 'categories', 'json_categories']]
df_processed.to_csv(out_path, index=False)

print(f"Đã xử lý và lưu file '{out_path}' thành công.")
print(df_processed.head()) # In vài dòng đầu của df_processed để kiểm tra
````

````python
import json
import pandas as pd
from collections import Counter, defaultdict

def analyze_arxiv_categories_extended(json_categories_data):
    """
    Phân tích json_categories:
    Trả về:
      - parent_count: Counter({parent: số_entry_chứa_parent})
      - parent_unique_subcats: dict({parent: set(nhãn_con_unique)})
      - parent_child_counts: dict({parent: Counter({child: tần_suất})})
      - total_entries: tổng số entry hợp lệ (parse JSON OK)
    """
    parent_count = Counter()
    parent_unique_subcats = defaultdict(set)
    parent_child_counts = defaultdict(Counter)
    total_entries = 0

    for line in json_categories_data:
        if not isinstance(line, (str, dict)) or (isinstance(line, str) and not line.strip()):
            continue

        try:
            categories = json.loads(line.strip()) if isinstance(line, str) else line
            if not isinstance(categories, dict):
                continue
            total_entries += 1

            for parent, subcats in categories.items():
                parent_count[parent] += 1
                # Chuẩn hoá subcats về list
                if subcats is None:
                    continue
                if not isinstance(subcats, list):
                    # Nếu là string "['a','b']" thì cố parse
                    try:
                        parsed = json.loads(subcats)
                        subcats = parsed if isinstance(parsed, list) else []
                    except Exception:
                        subcats = []

                for child in subcats:
                    parent_unique_subcats[parent].add(child)
                    parent_child_counts[parent][child] += 1

        except json.JSONDecodeError:
            # Bỏ qua JSON hỏng
            continue

    return parent_count, parent_unique_subcats, parent_child_counts, total_entries


def display_results_extended(parent_count, parent_unique_subcats, parent_child_counts, total_entries, top_k_children=10):
    """
    Hiển thị:
      - Tổng quan
      - Bảng nhãn cha: số lần xuất hiện + số nhãn con unique
      - Top-k nhãn con theo từng nhãn cha
    """
    total_parent_occurrences = sum(parent_count.values())
    print("📊 TỔNG QUAN")
    print("="*50)
    print(f"Tổng số entries hợp lệ: {total_entries}")
    print(f"Số nhãn cha duy nhất: {len(parent_count)}")
    print(f"Tổng lượt xuất hiện nhãn cha: {total_parent_occurrences}\n")

    # Bảng nhãn cha
    rows = []
    for p, c in parent_count.items():
        rows.append({
            "nhãn_cha": p,
            "số_lần_xuất_hiện": c,
            "số_nhãn_con_unique": len(parent_unique_subcats.get(p, set()))
        })
    df_parents = pd.DataFrame(rows).sort_values("số_lần_xuất_hiện", ascending=False)
    print("📋 THỐNG KÊ NHÃN CHA")
    print(df_parents.head(20).to_string(index=False))
    print()

    # Top-k nhãn con theo từng nhãn cha
    print("🔎 TOP nhãn con theo từng nhãn cha")
    print("="*50)
    for p in df_parents["nhãn_cha"].tolist():
        cnt = parent_child_counts.get(p, Counter())
        if not cnt:
            continue
        total_sub = sum(cnt.values())
        top_items = cnt.most_common(top_k_children)
        top_list = [f"{ch} ({n}, {n/total_sub*100:.1f}%)" for ch, n in top_items]
        print(f"- {p}: {'; '.join(top_list)}")
    print()

    # Thống kê tóm tắt
    print("📈 TÓM TẮT")
    print("="*50)
    for i, row in enumerate(df_parents.head(5).itertuples(index=False), 1):
        pct = row.số_lần_xuất_hiện / total_parent_occurrences * 100 if total_parent_occurrences else 0
        print(f"  {i}. {row.nhãn_cha}: {row.số_lần_xuất_hiện} lần ({pct:.1f}%)")


def results_to_dataframes(parent_count, parent_unique_subcats, parent_child_counts):
    """
    Chuyển kết quả sang 2 DataFrame:
      - df_parents: 1 dòng/nhãn_cha
      - df_parent_child: 1 dòng/(nhãn_cha, nhãn_con) với tần suất
    """
    df_parents = pd.DataFrame([
        {
            "nhãn_cha": p,
            "số_lần_xuất_hiện": parent_count[p],
            "số_nhãn_con_unique": len(parent_unique_subcats.get(p, set())),
            "danh_sách_nhãn_con_unique": ", ".join(sorted(parent_unique_subcats.get(p, set())))
        }
        for p in parent_count.keys()
    ]).sort_values("số_lần_xuất_hiện", ascending=False)

    rows = []
    for p, counter in parent_child_counts.items():
        for ch, n in counter.items():
            rows.append({"nhãn_cha": p, "nhãn_con": ch, "tần_suất": n})
    df_parent_child = pd.DataFrame(rows).sort_values(["nhãn_cha", "tần_suất"], ascending=[True, False])

    return df_parents, df_parent_child


def load_json_categories_column(file_path, col="json_categories"):
    """Đọc CSV và trả về list cột json_categories (bỏ NaN)."""
    df = pd.read_csv(file_path)
    if col not in df.columns:
        raise KeyError(f"Không tìm thấy cột '{col}' trong file.")
    return df[col].dropna().tolist()


# ============================
# Ví dụ chạy (điền lại file_path của bạn)
# ============================
if __name__ == "__main__":
    print("🚀 BẮT ĐẦU PHÂN TÍCH DỮ LIỆU ARXIV")
    file_path = "/content/drive/MyDrive/AIO/json_categories_only.csv"  # đổi theo đường dẫn của bạn
    data = load_json_categories_column(file_path)

    print("🔄 Đang phân tích dữ liệu...")
    parent_count, parent_unique_subcats, parent_child_counts, total_entries = analyze_arxiv_categories_extended(data)

    print("✅ Hoàn thành phân tích!\n")
    display_results_extended(parent_count, parent_unique_subcats, parent_child_counts, total_entries, top_k_children=None)

    # Xuất CSV nếu cần
    df_parents, df_parent_child = results_to_dataframes(parent_count, parent_unique_subcats, parent_child_counts)
    df_parents.to_csv("arxiv_parent_summary.csv", index=False, encoding="utf-8")
    df_parent_child.to_csv("arxiv_parent_child_counts.csv", index=False, encoding="utf-8")
    print("💾 Đã lưu: arxiv_parent_summary.csv & arxiv_parent_child_counts.csv")
````

Lấy top 50 nhãn con phổ biến nhất từ dữ liệu gốc (df_processed['categories']).

Tạo DataFrame mới gồm tối đa 300 dòng cho mỗi nhãn con, chọn theo thứ tự từ nhãn con nhiều → ít.

Khi xét nhãn con tiếp theo, đếm số lần nhãn đó đã xuất hiện trong tập đã chọn (do bị “ăn ké” từ các nhãn trước), và chỉ lấy bù phần thiếu từ các dòng chưa được lấy.

Yêu cầu dữ liệu gốc: df_processed có cột categories là list nhãn con (nếu là chuỗi dạng list cũng được, code sẽ chuẩn hoá).

````python
import pandas as pd
import ast
from collections import Counter, defaultdict

# --------------------------
# 1) Chuẩn hoá cột categories về List[str]
# --------------------------
def normalize_categories_col(df, col="categories"):
    if col not in df.columns:
        raise KeyError(f"Không tìm thấy cột '{col}' trong df.")
    def to_list(x):
        if isinstance(x, list):
            return x
        if pd.isna(x):
            return []
        if isinstance(x, str):
            # cố gắng parse kiểu "['a','b']" hoặc '["a","b"]'
            try:
                v = ast.literal_eval(x)
                return v if isinstance(v, list) else []
            except Exception:
                # fallback: tách theo dấu phẩy nếu cần
                return [s.strip() for s in x.split(",")] if "," in x else [x.strip()]
        return []
    out = df.copy()
    out[col] = out[col].apply(to_list)
    return out

# --------------------------
# 2) Tạo inverted index: child_label -> list chỉ số dòng
# --------------------------
def build_inverted_index(categories_series):
    inv = defaultdict(list)
    for i, cats in categories_series.items():
        for c in cats:
            inv[c].append(i)
    return inv

# --------------------------
# 3) Lấy top-K nhãn con theo tần suất xuất hiện (đếm theo số dòng có chứa nhãn đó)
# --------------------------
def top_k_child_labels(inverted_index, k=50):
    counts = {lbl: len(idxs) for lbl, idxs in inverted_index.items()}
    top = sorted(counts.items(), key=lambda x: x[1], reverse=True)[:k]
    return [lbl for lbl, _ in top], counts

# --------------------------
# 4) Chọn hàng theo chiến lược "điền quota 300/nhãn"
#     - Duy trì set chỉ số đã chọn
#     - Với mỗi nhãn: đếm đã có bao nhiêu trong set, bổ sung từ các ứng viên còn lại
# --------------------------
def select_rows_by_label_quota(df, inverted_index, labels_in_order, quota_per_label=300):
    selected = set()                         # chỉ số đã chọn
    label_indices_sets = {l: set(inverted_index.get(l, [])) for l in labels_in_order}
    achieved_counts = {}                     # nhãn -> số bản ghi thỏa chứa nhãn trong tập đã chọn

    for lbl in labels_in_order:
        lbl_set = label_indices_sets[lbl]
        already = len(selected & lbl_set)    # số dòng chứa nhãn này đã “có sẵn” trong tập
        missing = max(0, quota_per_label - already)

        if missing > 0:
            # ứng viên mới: các dòng chứa nhãn này nhưng chưa được chọn
            candidates = [i for i in inverted_index.get(lbl, []) if i not in selected]
            take = candidates[:missing]
            selected.update(take)

        # cập nhật lại achieved thực tế (sau khi thêm)
        achieved_counts[lbl] = len(selected & lbl_set)

    # Tạo df_out theo thứ tự xuất hiện ban đầu
    # (giữ nguyên thứ tự gốc để dễ kiểm tra)
    selected_sorted = sorted(list(selected))
    df_out = df.loc[selected_sorted].copy()

    return df_out, achieved_counts

# --------------------------
# 5) Pipeline chính
# --------------------------
def build_top50_quota_dataset(df_processed, categories_col="categories",
                              top_k=50, quota_per_label=300):
    # Chuẩn hoá
    df_norm = normalize_categories_col(df_processed, col=categories_col)

    # Inverted index
    inv = build_inverted_index(df_norm[categories_col])

    # Top-K nhãn con toàn bộ
    top_labels, global_counts = top_k_child_labels(inv, k=top_k)
    print(f"Top {top_k} nhãn con (theo số dòng chứa nhãn):")
    for i, lbl in enumerate(top_labels, 1):
        print(f"{i:>2}. {lbl}: {global_counts[lbl]} dòng")

    # Chọn theo quota
    print("\nĐang chọn dữ liệu theo quota mỗi nhãn =", quota_per_label)
    df_out, achieved = select_rows_by_label_quota(df_norm, inv, top_labels, quota_per_label=quota_per_label)

    # Thống kê đạt được
    print(f"\n✅ Hoàn tất. Số dòng trong tập kết quả: {len(df_out)}")
    print("Số dòng đạt được cho từng nhãn (đã tính cả 'ăn ké' từ nhãn trước):")
    for lbl in top_labels:
        got = achieved.get(lbl, 0)
        note = "" if got >= quota_per_label else f" (thiếu {quota_per_label - got})"
        print(f"- {lbl}: {got}{note}")

    return df_out, top_labels, achieved

# ============================
# CÁCH DÙNG:
# df_processed: DataFrame gốc, có cột 'categories'
# ============================

# df_selected, top50_labels, per_label_counts = build_top50_quota_dataset(
#     df_processed,
#     categories_col="categories",
#     top_k=50,
#     quota_per_label=300
# )

# # (tuỳ chọn) lưu ra CSV
# df_selected.to_csv("top50_child_labels_quota300.csv", index=False, encoding="utf-8")
````

````python
import ast

# chuẩn hoá categories thành list
def normalize_categories(x):
    if isinstance(x, list):
        return x
    if pd.isna(x):
        return []
    if isinstance(x, str):
        try:
            v = ast.literal_eval(x)
            return v if isinstance(v, list) else [x.strip()]
        except Exception:
            return [s.strip() for s in x.split(",")] if "," in x else [x.strip()]
    return []

# giả sử bạn có list 50 labels trong biến list50labels
# ví dụ: list50labels = ["hep-th", "hep-ph", "gr-qc", "quant-ph", ...]

# Tạo dataframe mới từ df_selected mà không tác động đến bản gốc
df_selected_top50 = df_selected.copy()
df_selected_top50["categories"] = df_selected_top50["categories"].apply(normalize_categories)

# lọc lại categories theo list50labels
df_selected_top50["categories"] = df_selected_top50["categories"].apply(
    lambda cats: [c for c in cats if c in list50labels]
)

# loại bỏ các dòng trống sau khi lọc
df_selected_top50 = df_selected_top50[df_selected_top50["categories"].map(len) > 0].reset_index(drop=True)

print("✅ Hoàn tất. Số dòng còn lại:", len(df_selected_top50))
print(df_selected_top50.head())
````

````python
import pandas as pd
import numpy as np
import ast
import time

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.metrics import f1_score, accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sentence_transformers import SentenceTransformer

# ----------------- 1. LOAD DATA -----------------
df = df_selected_top50.copy()

# Ghép 3 cột authors + title + abstract thành text
df["text"] = (
    df["authors"].astype(str) + " " +
    df["title"].astype(str) + " " +
    df["abstract"].astype(str)
)

texts = df["text"].tolist()
labels = df["categories"].tolist()
print(f"✅ Data: {len(df)} samples | text + categories ready")

# ----------------- 2. LABELS -----------------
mlb = MultiLabelBinarizer()
Y = mlb.fit_transform(labels)
print(f"✅ {len(mlb.classes_)} unique labels. Ví dụ: {mlb.classes_[:10]}")

# ----------------- 3. EMBEDDINGS -----------------
print("🔄 Đang tạo embedding với SentenceTransformer...")
start = time.time()
model = SentenceTransformer("all-MiniLM-L6-v2", device="cuda")
X = model.encode(texts, show_progress_bar=True, device="cuda")
X = np.array(X, dtype="float32")
print(f"✅ Embedding xong: {X.shape} (thời gian {time.time()-start:.2f}s)")


````

````python

from sklearn.metrics import classification_report, jaccard_score
from iterstrat.ml_stratifiers import MultilabelStratifiedKFold, MultilabelStratifiedShuffleSplit

# ----------------- 4. TRAIN/TEST SPLIT (MULTILABEL STRATIFY) -----------------
msss = MultilabelStratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_idx, test_idx in msss.split(X, Y):
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = Y[train_idx], Y[test_idx]
    idx_train, idx_test = df.index[train_idx], df.index[test_idx]

print(f"✅ Split: train={X_train.shape[0]}, test={X_test.shape[0]}")

# ----------------- 5. TRAIN KNN -----------------
print("🔄 Huấn luyện KNN...")
knn = KNeighborsClassifier(n_neighbors=5, metric="cosine")
start = time.time()
knn.fit(X_train, y_train)
print(f"✅ Train xong (thời gian {time.time()-start:.2f}s)")

# ----------------- 6. DỰ ĐOÁN -----------------
y_pred = knn.predict(X_test)
acc = accuracy_score(y_test, y_pred)
f1_micro = f1_score(y_test, y_pred, average="micro")
f1_macro = f1_score(y_test, y_pred, average="macro")
jaccard = jaccard_score(y_test, y_pred, average="samples")  # dùng trung bình theo từng sample

print(f"🎯 Đánh giá KNN → Accuracy={acc:.4f}, F1-micro={f1_micro:.4f}, "
      f"F1-macro={f1_macro:.4f}, Jaccard={jaccard:.4f}")

# ----------------- 7. TẠO predict_df -----------------
true_labels = mlb.inverse_transform(y_test)
pred_labels = mlb.inverse_transform(y_pred)

predict_df = df.loc[idx_test].copy()
predict_df["true_labels"] = true_labels
predict_df["predicted_labels"] = pred_labels

print("📊 Ví dụ 5 sample đầu tiên trong predict_df:")
print(predict_df[["authors", "title", "true_labels", "predicted_labels"]].head())

# ----------------- 8. SAVE KẾT QUẢ -----------------
predict_df.to_csv("/content/drive/MyDrive/AIO/predict_results.csv", index=False, encoding="utf-8")
print("💾 Đã lưu kết quả dự đoán vào file: /content/drive/MyDrive/AIO/predict_results.csv")

# ----------------- 9. REPORT CHI TIẾT -----------------
print("\n📊 Classification Report chi tiết cho từng nhãn:")
report = classification_report(y_test, y_pred, target_names=mlb.classes_, zero_division=0)
print(report)
````

# 📑 Report Kết Quả Mô Hình KNN + ANN

## 1. Ý nghĩa các metrics

### 🔹 Accuracy

* Trong **multi-label classification**, một sample chỉ được coi là đúng khi **toàn bộ tập nhãn** khớp với ground truth.
* Do đó Accuracy thường thấp hơn so với F1.

### 🔹 F1-micro

* Tính **TP, FP, FN trên toàn bộ dataset** (gộp hết các nhãn).
* **Ưu tiên nhãn phổ biến** (nhiều dữ liệu).
* Đo mức độ mô hình làm tốt trên toàn cục.

### 🔹 F1-macro

* Tính **F1 riêng cho từng nhãn**, sau đó lấy trung bình cộng.
* **Tất cả nhãn được coi trọng như nhau** (cân bằng giữa nhãn nhiều và nhãn ít).
* Đo mức độ mô hình có **công bằng với các nhãn nhỏ/lẻ** hay không.
