```python

# Code t·ªëi ∆∞u nh·∫•t - In 100 categories ph·ªï bi·∫øn nh·∫•t
print("=== TH·ªêNG K√ä 100 CATEGORIES PH·ªî BI·∫æN NH·∫§T ===")

import pandas as pd
import time

# B·∫Øt ƒë·∫ßu ƒëo th·ªùi gian
start_time = time.time()

print("üîÑ ƒêang chuy·ªÉn dataset sang pandas...")

# Chuy·ªÉn dataset sang pandas (t·ªëi ∆∞u cho 32GB RAM)
df = full_dataset['train'].to_pandas()

print(f"‚úÖ Ho√†n th√†nh trong {time.time() - start_time:.1f} gi√¢y!")

# Th·ªëng k√™ categories (ch·ªâ l·∫•y top 100)
category_stats = df['categories'].value_counts().head(100)

# Hi·ªÉn th·ªã k·∫øt qu·∫£
print(f"\nÔøΩÔøΩ TOP 100 CATEGORIES PH·ªî BI·∫æN NH·∫§T:")
print(f"T·ªïng s·ªë b√†i b√°o: {len(df):,}")
print(f"T·ªïng th·ªùi gian: {time.time() - start_time:.1f} gi√¢y")
print("-" * 70)

# In ra 100 categories
for i, (category, count) in enumerate(category_stats.items(), 1):
    percentage = (count / len(df)) * 100
    print(f"{i:3d}. {category:<30} | {count:>8,} b√†i ({percentage:5.2f}%)")
```

```python
# ===== L√ÄM S·∫†CH CATEGORIES - LO·∫†I B·ªé PH·∫¶N SAU D·∫§U "." =====
print("=== L√ÄM S·∫†CH CATEGORIES ===")

import pandas as pd
import time

# B·∫Øt ƒë·∫ßu ƒëo th·ªùi gian
start_time = time.time()

print("üîÑ ƒêang l√†m s·∫°ch categories...")

# T·∫°o b·∫£n sao c·ªßa dataframe ƒë·ªÉ kh√¥ng ·∫£nh h∆∞·ªüng ƒë·∫øn d·ªØ li·ªáu g·ªëc
df_cleaned = df.copy()

# Hi·ªÉn th·ªã m·ªôt v√†i v√≠ d·ª• tr∆∞·ªõc khi l√†m s·∫°ch
print("\nüìã V√ç D·ª§ TR∆Ø·ªöC KHI L√ÄM S·∫†CH:")
sample_categories = df_cleaned['categories'].head(10).tolist()
for i, cat in enumerate(sample_categories, 1):
    print(f"  {i:2d}. {cat}")

# Lo·∫°i b·ªè ph·∫ßn sau d·∫•u "." trong c·ªôt categories
# S·ª≠ d·ª•ng str.split('.') ƒë·ªÉ t√°ch chu·ªói t·∫°i d·∫•u "."
# Sau ƒë√≥ l·∫•y ph·∫ßn ƒë·∫ßu ti√™n [0] (tr∆∞·ªõc d·∫•u ".")
df_cleaned['categories_cleaned'] = df_cleaned['categories'].str.split('.').str[0]

# Hi·ªÉn th·ªã m·ªôt v√†i v√≠ d·ª• sau khi l√†m s·∫°ch
print("\n‚ú® V√ç D·ª§ SAU KHI L√ÄM S·∫†CH:")
sample_cleaned = df_cleaned['categories_cleaned'].head(10).tolist()
for i, cat in enumerate(sample_cleaned, 1):
    print(f"  {i:2d}. {cat}")

print(f"\n‚úÖ Ho√†n th√†nh l√†m s·∫°ch trong {time.time() - start_time:.1f} gi√¢y!")
```

```python
# ===== TH·ªêNG K√ä CATEGORIES SAU KHI L√ÄM S·∫†CH =====
print("=== TH·ªêNG K√ä CATEGORIES SAU KHI L√ÄM S·∫†CH ===")

# Th·ªëng k√™ categories ƒë√£ l√†m s·∫°ch (l·∫•y top 50)
cleaned_category_stats = df_cleaned['categories_cleaned'].value_counts().head(1000)

# Hi·ªÉn th·ªã k·∫øt qu·∫£
print(f"\nüìä TOP 50 CATEGORIES SAU KHI L√ÄM S·∫†CH:")
print(f"T·ªïng s·ªë b√†i b√°o: {len(df_cleaned):,}")
print(f"T·ªïng th·ªùi gian: {time.time() - start_time:.1f} gi√¢y")
print("-" * 70)

# In ra 50 categories ƒë√£ l√†m s·∫°ch
for i, (category, count) in enumerate(cleaned_category_stats.items(), 1):
    percentage = (count / len(df_cleaned)) * 100
    print(f"{i:2d}. {category:<20} | {count:>8,} b√†i ({percentage:5.2f}%)")

# So s√°nh s·ªë l∆∞·ª£ng categories tr∆∞·ªõc v√† sau khi l√†m s·∫°ch
original_unique = df['categories'].nunique()
cleaned_unique = df_cleaned['categories_cleaned'].nunique()

print(f"\nüìà SO S√ÅNH:")
print(f"  - S·ªë categories g·ªëc: {original_unique:,}")
print(f"  - S·ªë categories sau l√†m s·∫°ch: {cleaned_unique:,}")
print(f"  - Gi·∫£m: {original_unique - cleaned_unique:,} categories ({((original_unique - cleaned_unique) / original_unique * 100):.1f}%)")
```

```python
# ===== L·ªåC D·ªÆ LI·ªÜU THEO TOP 40 CATEGORIES =====
print("=== L·ªåC D·ªÆ LI·ªÜU THEO TOP 40 CATEGORIES ===")

import pandas as pd
import time

# B·∫Øt ƒë·∫ßu ƒëo th·ªùi gian
start_time = time.time()

print("üîÑ ƒêang l·ªçc d·ªØ li·ªáu theo top 40 categories...")

# 1. L·∫•y top 40 categories ph·ªï bi·∫øn nh·∫•t (sau khi l√†m s·∫°ch)
top_40_categories = df_cleaned['categories_cleaned'].value_counts().head(40)

print(f"\nüìä TOP 40 CATEGORIES ƒê∆Ø·ª¢C CH·ªåN:")
print("H·∫°ng | Category           | S·ªë b√†i b√°o | T·ª∑ l·ªá")
print("-" * 55)

for i, (category, count) in enumerate(top_40_categories.items(), 1):
    percentage = (count / len(df_cleaned)) * 100
    print(f"{i:3d}  | {category:<18} | {count:>8,} | {percentage:5.2f}%")

# 2. L·ªçc dataset ch·ªâ gi·ªØ l·∫°i c√°c b√†i b√°o thu·ªôc top 40 categories
print(f"\nüîç ƒêang l·ªçc dataset...")

# T·∫°o mask ƒë·ªÉ l·ªçc d·ªØ li·ªáu
mask = df_cleaned['categories_cleaned'].isin(top_40_categories.index)

# T·∫°o dataset m·ªõi
df_top40 = df_cleaned[mask].copy()

print(f"‚úÖ Ho√†n th√†nh l·ªçc d·ªØ li·ªáu trong {time.time() - start_time:.1f} gi√¢y!")

# 3. Th·ªëng k√™ dataset m·ªõi
print(f"\nüìà TH·ªêNG K√ä DATASET M·ªöI:")
print(f"  - S·ªë b√†i b√°o g·ªëc: {len(df_cleaned):,}")
print(f"  - S·ªë b√†i b√°o sau l·ªçc: {len(df_top40):,}")
print(f"  - T·ª∑ l·ªá gi·ªØ l·∫°i: {(len(df_top40)/len(df_cleaned)*100):.1f}%")
print(f"  - S·ªë categories: {df_top40['categories_cleaned'].nunique()}")

# 4. Ki·ªÉm tra ph√¢n b·ªë d·ªØ li·ªáu
print(f"\nüìä PH√ÇN B·ªê D·ªÆ LI·ªÜU TRONG DATASET M·ªöI:")
category_distribution = df_top40['categories_cleaned'].value_counts()

print("H·∫°ng | Category           | S·ªë b√†i b√°o | T·ª∑ l·ªá trong dataset m·ªõi")
print("-" * 70)

for i, (category, count) in enumerate(category_distribution.items(), 1):
    percentage = (count / len(df_top40)) * 100
    print(f"{i:3d}  | {category:<18} | {count:>8,} | {percentage:5.2f}%")
```

```python
# ===== KI·ªÇM TRA CH·∫§T L∆Ø·ª¢NG DATASET M·ªöI =====
print("=== KI·ªÇM TRA CH·∫§T L∆Ø·ª¢NG DATASET M·ªöI ===")

# 1. Ki·ªÉm tra c·∫•u tr√∫c dataset m·ªõi
print("üîç 1. C·∫§U TR√öC DATASET M·ªöI:")
print(f"  - K√≠ch th∆∞·ªõc: {df_top40.shape}")
print(f"  - C√°c c·ªôt: {list(df_top40.columns)}")

# 2. Ki·ªÉm tra d·ªØ li·ªáu thi·∫øu
print(f"\nüîç 2. KI·ªÇM TRA D·ªÆ LI·ªÜU THI·∫æU:")
missing_data = df_top40.isnull().sum()
for col, missing_count in missing_data.items():
    if missing_count > 0:
        percentage = (missing_count / len(df_top40)) * 100
        print(f"  - {col}: {missing_count:,} ({percentage:.2f}%)")

# 3. Hi·ªÉn th·ªã m·ªôt v√†i v√≠ d·ª• d·ªØ li·ªáu
print(f"\nüîç 3. V√ç D·ª§ D·ªÆ LI·ªÜU TRONG DATASET M·ªöI:")
print("Index | Categories (g·ªëc)     | Categories (cleaned) | Title (50 k√Ω t·ª± ƒë·∫ßu)")
print("-" * 90)

# L·∫•y 10 v√≠ d·ª• ƒë·∫ßu ti√™n
sample_data = df_top40[['categories', 'categories_cleaned', 'title']].head(10)
for idx, row in sample_data.iterrows():
    title_short = row['title'][:50] + "..." if len(row['title']) > 50 else row['title']
    print(f"{idx:5d} | {row['categories']:<20} | {row['categories_cleaned']:<18} | {title_short}")

# 4. Th·ªëng k√™ ƒë·ªô d√†i abstract
print(f"\nüîç 4. TH·ªêNG K√ä ƒê·ªò D√ÄI ABSTRACT:")
if 'abstract' in df_top40.columns:
    abstract_lengths = df_top40['abstract'].str.len()
    print(f"  - ƒê·ªô d√†i trung b√¨nh: {abstract_lengths.mean():.0f} k√Ω t·ª±")
    print(f"  - ƒê·ªô d√†i ng·∫Øn nh·∫•t: {abstract_lengths.min():,} k√Ω t·ª±")
    print(f"  - ƒê·ªô d√†i d√†i nh·∫•t: {abstract_lengths.max():,} k√Ω t·ª±")
    print(f"  - ƒê·ªô l·ªách chu·∫©n: {abstract_lengths.std():.0f} k√Ω t·ª±")

# 5. L∆∞u dataset m·ªõi
print(f"\nüíæ 5. L∆ØU DATASET M·ªöI:")
print("  - Dataset m·ªõi ƒë∆∞·ª£c l∆∞u trong bi·∫øn 'df_top40'")
print("  - B·∫°n c√≥ th·ªÉ s·ª≠ d·ª•ng df_top40 cho c√°c b∆∞·ªõc ti·∫øp theo")
print("  - K√≠ch th∆∞·ªõc nh·ªè h∆°n, d·ªÖ x·ª≠ l√Ω h∆°n cho machine learning")
```

```python
# ===== L∆ØU DATASET M·ªöI V√ÄO FILE =====
print("=== L∆ØU DATASET M·ªöI V√ÄO FILE ===")

import os

# 1. T·∫°o th∆∞ m·ª•c ƒë·ªÉ l∆∞u dataset m·ªõi
output_dir = "filtered_datasets"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
    print(f"‚úÖ ƒê√£ t·∫°o th∆∞ m·ª•c: {output_dir}")

# 2. L∆∞u dataset m·ªõi d∆∞·ªõi d·∫°ng CSV
output_file = os.path.join(output_dir, "arxiv_top40_categories.csv")
print(f"üîÑ ƒêang l∆∞u dataset v√†o: {output_file}")

# L∆∞u file CSV
df_top40.to_csv(output_file, index=False, encoding='utf-8')

print(f"‚úÖ ƒê√£ l∆∞u dataset th√†nh c√¥ng!")
print(f"  - File: {output_file}")
print(f"  - K√≠ch th∆∞·ªõc: {df_top40.shape}")
print(f"  - Dung l∆∞·ª£ng file: {os.path.getsize(output_file) / (1024*1024):.1f} MB")

# 3. T·∫°o file th·ªëng k√™
stats_file = os.path.join(output_dir, "dataset_statistics.txt")
print(f"\nüìä ƒêang t·∫°o file th·ªëng k√™: {stats_file}")

with open(stats_file, 'w', encoding='utf-8') as f:
    f.write("=== TH·ªêNG K√ä DATASET TOP 40 CATEGORIES ===\n\n")
    f.write(f"T·ªïng s·ªë b√†i b√°o: {len(df_top40):,}\n")
    f.write(f"S·ªë categories: {df_top40['categories_cleaned'].nunique()}\n")
    f.write(f"T·ª∑ l·ªá gi·ªØ l·∫°i t·ª´ dataset g·ªëc: {(len(df_top40)/len(df_cleaned)*100):.1f}%\n\n")
    
    f.write("TOP 40 CATEGORIES:\n")
    f.write("-" * 50 + "\n")
    for i, (category, count) in enumerate(category_distribution.items(), 1):
        percentage = (count / len(df_top40)) * 100
        f.write(f"{i:2d}. {category:<18} | {count:>8,} b√†i ({percentage:5.2f}%)\n")

print(f"‚úÖ ƒê√£ t·∫°o file th·ªëng k√™ th√†nh c√¥ng!")

# 4. Hi·ªÉn th·ªã th√¥ng tin cu·ªëi c√πng
print(f"\nüéâ HO√ÄN TH√ÄNH L·ªåC D·ªÆ LI·ªÜU!")
print("=" * 50)
print(f"üìÅ Th∆∞ m·ª•c output: {output_dir}")
print(f"üìÑ File d·ªØ li·ªáu: arxiv_top40_categories.csv")
print(f"üìä File th·ªëng k√™: dataset_statistics.txt")
print(f"üî¢ S·ªë b√†i b√°o: {len(df_top40):,}")
print(f"üè∑Ô∏è  S·ªë categories: {df_top40['categories_cleaned'].nunique()}")
print(f"üíæ Bi·∫øn trong notebook: df_top40")
```
