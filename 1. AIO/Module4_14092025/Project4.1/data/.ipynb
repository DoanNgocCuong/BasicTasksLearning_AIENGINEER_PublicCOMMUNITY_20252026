```python
import pandas as pd
import ast
import json

# ===== Cấu hình =====
data_path = "arxiv_perfectly_balanced.csv"  # đổi nếu cần
out_path  = "dropped_col_df.csv"
cols = ["authors", "title", "abstract", "categories", "parent_labels", "child_labels"]

# ===== Đọc file chỉ lấy cột cần =====
df = pd.read_csv(data_path, usecols=cols)

# ===== Chuẩn hoá 2 cột list (nếu đang lưu dạng chuỗi list trong CSV) =====
for col in ["parent_labels", "child_labels"]:
    df[col] = df[col].apply(
        lambda x: x if isinstance(x, list)
        else ([] if pd.isna(x) else ast.literal_eval(str(x)))
    )

# ===== Tạo cột json_categories: với mỗi cha, quét toàn bộ con để lấy con có tiền tố đúng =====
df["json_categories"] = df.apply(
    lambda r: {
        p: [c for c in (r["child_labels"] or [])
             if isinstance(c, str) and (c == p or c.startswith(p + ".") or c.startswith(p + "-"))]
        for p in (r["parent_labels"] or [])
    },
    axis=1
)

# Tuỳ chọn: nếu muốn lưu ra CSV ở dạng chuỗi JSON cho cột json_categories
df["json_categories"] = df["json_categories"].apply(lambda d: json.dumps(d, ensure_ascii=False))

# ===== Lưu kết quả =====
df_processed=df.drop(columns=["categories", "parent_labels", "child_labels"])
df_processed.to_csv(out_path, index=False)  # bỏ tham số file_name khỏi đoạn lệnh to_csv
```
